# Most like part of PT

```python
import pandas as pd
import string
import nltk
import sklearn
import numpy as np
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.corpus import stopwords
from textblob import TextBlob
from textblob import Word
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
like = pd.read_excel('/Users/lucida/Desktop/nlp project/data/like.xlsx', sheet_name = 'Sheet1')
orig_like = pd.read_excel('/Users/lucida/Desktop/nlp project/data/like.xlsx', sheet_name = 'Sheet1')
```

```python
like.head
```

## Preprocessing

```python
#convert to lower case
like['most_like'] = like['most_like'].apply(lambda x: " ".join(x.lower() for x in x.split()))
#remove punctuation
like['most_like'] = like['most_like'].str.replace('[{}]'.format(string.punctuation), '')
#remove whightspace
like['most_like'] = like['most_like'].str.strip()
```

```python
#remove stop words
stop = nltk.corpus.stopwords.words('english')
#stop.append(['interesting', 'good', 'love', 'like', 'really'])
like['most_like'] = like['most_like'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
#spelling correction
#note: can be slow for large data
like['most_like'] = like['most_like'].apply(lambda x: str(TextBlob(x).correct()))
```

```python
#lemmatization
from textblob import Word
like['most_like'] = like['most_like'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```python
stop.append(['interesting', 'good', 'love', 'like', 'really', 'great'])
like['most_like'] = like['most_like'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

## WordCloud

```python
text = " ".join(review for review in like.most_like)
print ("There are {} words in the combination of all review.".format(len(text)))
# Create stopword list:
stopwords = set(STOPWORDS)

stopwords.update(['interesting', 'good', 'article', 'physic', 'field','science', 'topic', 'issue', 'recent', 'read', 'interest', 'great', 'reading', 'covered', 'coverage','area'])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white"
                      , min_font_size = 9).generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## word matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=100, analyzer='word',
 stop_words= 'english',ngram_range=(1,1))
train_vect = tfidf.fit_transform(like['most_like'])
train_vect.shape
```

## cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
simi_2 = cosine_similarity(train_vect[1:2], train_vect)
```

## hierarchical clustering

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from scipy.cluster import  hierarchy

#Vectorizing
X = CountVectorizer().fit_transform(like['most_like'])
X = TfidfTransformer().fit_transform(X)
#Clustering
X = X.todense()
threshold = 0.1
Z = hierarchy.linkage(X,"average", metric="cosine")
```

```python
C = hierarchy.fcluster(Z, t=0.95, criterion='distance')
#C.sort()
```

```python
#visulization
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
# Specify that we want a figure with a width of 15 inches and a height of 8 inches
fig = plt.figure(figsize=(100, 20))

# Title of the dendrogram
plt.title('like - Agglomerative clustering - average linkage', fontsize=20)

# Clearly specify what is plotted on the y-axis
plt.ylabel('Cluster distance', fontsize=20)

dn = dendrogram(
    # the Z-matrix for which we want to create a dendrogram
    Z=Z,
    # orientation of the dendrogram
    orientation='top',
    # labels for the objects, if available
  
    # font-size for the object labels
    leaf_font_size=16,
    # rotation of the object labels
    leaf_rotation=90,
)
```

```python
orig_like['cluster'] = C
```

```python
cluster_review = orig_like.sort_values(by=['cluster'])
```

```python
cluster_review.to_excel("cluster_review_like2.xlsx")
```

```python
series_like = like.most_like.str.split(expand=True).stack().value_counts()
```

```python
opened_file = open('source2.csv')
read_file = reader(opened_file)
list_source = list(read_file)
list_source = list_source[1:]
```

```python
list_source[0][1] = list_source[0][1].replace('time', 'NYT')
```

```python
#clean New York Times
nyt = ['ny times', 'new york times']
def clean_NYT(string):
    for char in nyt:
        string = string.replace(char,"NYT")
    return string

for row in list_source:
    review = row[1]
    review = clean_NYT(review)
    row[1] = review
```

```python
#clean Sky & Telescope
sky = ['sky & telescope', 'sky and telescope' ]
def clean_SKY(string):
    for char in sky:
        string = string.replace(char,"SKY")
    return string

for row in list_source:
    review = row[1]
    review = clean_SKY(review)
    row[1] = review
```

```python
#clean American Scientist
for row in list_source:
    review = row[1]
    review = review.replace('american scientist', 'AmericanSci')
    row[1] = review
```
